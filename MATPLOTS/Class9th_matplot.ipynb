# **LINE PLOT**

```python
# Define Limit of X and Y cordinates and Plot a line on it

xpoint = np.array([0,6])
ypoint = np.array([0,250])
#Store in memory
plt.plot(xpoint,ypoint)
#Show the graph
plt.show
```

# **FORMAT STRING**

```python
xpoint = np.array([1,4,6,8])

plt.plot(xpoint,'*-.r')
plt.show

# : --- for doted line
# - = for solid line
# -- = for dashed line
# -. = for dashed doted line 
```

### Create Labels for a Plot

```python
x=np.array([1,4,6,8])
y=np.array([5,6,8,9])

plt.plot(x,y)
plt.xlabel("AVERAGE NUMBER")
plt.ylabel("TOTAL POINT")
plt.show
```

### Creta a Title

```python
x=np.array([1,4,6,8])
y=np.array([5,6,8,9])

plt.plot(x,y)
plt.xlabel("AVERAGE NUMBER")
plt.ylabel("TOTAL POINT")
plt.title("HUMAN VALUES")
plt.show
```

### Edit Fonts

```python
x=np.array([1,4,6,8])
y=np.array([5,6,8,9])

fo={'family':'serif','color':'blue','size':20}
ft={'family':'serif','color':'hotpink','size':20}

plt.plot(x,y)
plt.xlabel("AVERAGE NUMBER",fontdict=fo)
plt.ylabel("TOTAL POINT",fontdict=fo)
plt.title("HUMAN VALUES",fontdict=ft)
plt.show
```

### Position Title

```python
x=np.array([1,4,6,8])
y=np.array([5,6,8,9])

fo={'family':'serif','color':'blue','size':20}
ft={'family':'serif','color':'hotpink','size':20}

plt.plot(x,y)
plt.xlabel("AVERAGE NUMBER",fontdict=fo)
plt.ylabel("TOTAL POINT",fontdict=fo)
plt.title("HUMAN VALUES",fontdict=ft,loc='left')
plt.show
```

### ColourMap

```python
# VIRIDIS
x=np.array([1,4,6,8,2,6,7,8,5])
y=np.array([5,6,8,9,11,22,33,12,45])

c=np.array([0,11,23,34,45,56,78,89,99])

plt.scatter(x,y,c=c,cmap='viridis')

plt.show
```

### ColourBar

```python
# VIRIDIS
x=np.array([1,4,6,8,2,6,7,8,5])
y=np.array([5,6,8,9,11,22,33,12,45])

c=np.array([0,11,23,34,45,56,78,89,99])

plt.scatter(x,y,c=c,cmap='viridis')
plt.colorbar()

plt.show
```

```python
# VIRIDIS
x=np.array([1,4,6,8,2,6,7,8,5])
y=np.array([5,6,8,9,11,22,33,12,45])

s=np.array([0,11,23,34,45,56,78,89,99])

plt.scatter(x,y,s=s)


plt.show
```

```python
# ALPHA -- Transparent
x=np.array([1,4,6,8,2,6,7,8,5])
y=np.array([5,6,8,9,11,22,33,12,45])

s=np.array([0,11,23,34,45,56,78,89,99])

plt.scatter(x,y,s=s,alpha=0.5)


plt.show
```

```python
# COMBINE COLOR SIZE
# ALPHA -- Transparent
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
c=np.random.randint(100,size=(100))
s=10*np.random.randint(100,size=(100))

plt.scatter(x,y,c=c,s=s,alpha=0.9,cmap='nipy_spectral')
plt.colorbar()
plt.show
```

# PIE CHART

```python
x=np.array([20,20,20,20,20])

plt.pie(x)
plt.show
```

### LABELS

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])

plt.pie(x,labels=y)
plt.show
```

### SET ANGLE

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])

plt.pie(x,labels=y,startangle=45)
plt.show
```

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])

plt.pie(x,labels=y,startangle=90)
plt.show
```

### EXPLODE

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])
e=[1,0,0,0,0]

plt.pie(x,labels=y,startangle=90,explode=e)
plt.show
```

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])
e=[0.1,0.2,0.3,0.4,0.4]

plt.pie(x,labels=y,startangle=90,explode=e)
plt.show
```

### Shadow

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])
e=[0.2,0,0,0,0]

plt.pie(x,labels=y,startangle=90,explode=e,shadow=True)
plt.show
```

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])
e=[0.2,0,0,0,0]
z=["red","yellow","purple","brown","yellow"]

plt.pie(x,labels=y,startangle=90,explode=e,shadow=True,colors=z)
plt.show
```

### Legend

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])
e=[0.2,0,0,0,0]
z=["red","yellow","purple","brown","yellow"]

plt.pie(x,labels=y,startangle=90,explode=e,shadow=True,colors=z)
plt.legend()
plt.show
```

```python
x=np.array([20,20,20,20,20])
y=np.array(["APPLE","BANANAS","CHERRYS","NUTS","MANGO"])
e=[0.2,0,0,0,0]
z=["red","yellow","purple","brown","yellow"]

plt.pie(x,labels=y,startangle=90,explode=e,shadow=True,colors=z)
plt.legend(title="FRUITS")
plt.show
```
